```python
import pandas as pd
import seaborn as sns
import matplotlib as mp
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv("./data/house_pricing.csv", sep=",")
print("Dataset loaded")
```

```python
print(dataset.dtypes)
```

```python
dfprice = dataset["SalePrice"]
dfarea = dataset["LotArea"]

sns.scatterplot(dfprice, dfarea)
```

```python
df = dataset[(dataset['LotArea'] < 20000) & (df['SalePrice'] < 500000)]
dfprice = df['SalePrice']
dfarea = df['LotArea']

sns.scatterplot(dfarea, dfprice)
```

```python
dfprice = dataset["LotFrontage"]
dfarea = dataset["LotArea"]

sns.scatterplot(dfprice, dfarea)
```

```python
df = dataset[(dataset['LotFrontage'] < 200) & (df['LotArea'] < 100000)]
dffrontage = df['LotFrontage']
dfarea = df['LotArea']

sns.scatterplot(dffrontage, dfarea)
```

```python
dataset = pd.read_csv("./data/IBM_HR_ATTRITION.csv", sep=",")
print("Dataset loaded")
```

```python
dataset.describe()
```

```python
x = dataset['JobSatisfaction']
y = dataset['Attrition']

g = sns.catplot(x = x,y = y, data=dataset)
```

```python
x = dataset['JobSatisfaction']
y = dataset['Attrition']

g = sns.boxplot(x = x,y = y, data=dataset)
```

```python
dataset.describe
```

```python
x = dataset["JobSatisfaction"]
y = dataset["TotalWorkingYears"]

sns.boxplot(x=x, y=y, data=dataset)
```

```python
df = dataset[dataset['Attrition'] == 'Yes']
x = df["EducationField"]
y = df["JobSatisfaction"]

sns.barplot(x=x, y=y, data=df)
```

```python
d = dataset["MonthlyIncome"]

sns.distplot(d)
```

```python
x = dataset[dataset["MonthlyIncome"] < 5000]

sns.distplot(x)
```

```python
y = dataset["Attrition"].replace('Yes',1).replace('No',0)
x = dataset["DistanceFromHome"]

sns.lmplot(x=x, y=y, data=dataset)
```

```python
y = dataset["Attrition"].replace('Yes',1).replace('No',0)
x = dataset["YearsWithCurrManager"]

sns.lmplot(x=x, y=y, data=dataset)
```
