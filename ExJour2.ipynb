```python
import pandas as pd
import seaborn as sns
import matplotlib as mp
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv("./data/house_pricing.csv", sep=",")
print("Dataset loaded")
```

```python
print(dataset.dtypes)
```

```python
dfprice = dataset["SalePrice"]
dfarea = dataset["LotArea"]

sns.scatterplot(dfprice, dfarea)
```

```python
df = dataset[(dataset['LotArea'] < 20000) & (df['SalePrice'] < 500000)]
dfprice = df['SalePrice']
dfarea = df['LotArea']

sns.scatterplot(dfprice, dfarea)
```

```python
dfprice = dataset["LotFrontage"]
dfarea = dataset["LotArea"]

sns.scatterplot(dfprice, dfarea)
```

```python
df = dataset[(dataset['LotFrontage'] < 200) & (df['LotArea'] < 100000)]
dffrontage = df['LotFrontage']
dfarea = df['LotArea']

sns.scatterplot(dffrontage, dfarea)
```
